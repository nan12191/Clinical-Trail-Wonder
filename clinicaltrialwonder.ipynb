This is an attempt to mine clinical trials data

```python
import requests
import json
import pandas as pd
import datetime
```

```python
#from google.colab import drive
#drive.mount('/content/drive')
datafolder = '/content/drive/MyDrive/ClinicalTrialWonder'
```

```python
immunotherapy_BASE_URL = 'https://clinicaltrials.gov/api/query/study_fields?expr="immune checkpoint inhibitors"&max_rnk=1000&fmt=json'

cdc_extract_fields = [
    'BriefTitle',
    'DesignAllocation',
    'DesignMasking',
    'DesignMaskingDescription',
    'InterventionName',
    'InterventionType',
    'LastKnownStatus',
    'OfficialTitle',
    'OutcomeAnalysisStatisticalMethod',
    'OutcomeMeasureTimeFrame',
    'SecondaryOutcomeMeasure',
    'StartDate',
    'StudyFirstPostDate',
    'StudyFirstPostDateType',
    'StudyFirstSubmitDate',
    'StudyFirstSubmitQCDate',
    'StudyPopulation',
    'StudyType',
    'WhyStopped'
]

query_url = f'{immunotherapy_BASE_URL}&fields={",".join(cdc_extract_fields)}'
print(query_url)

r = requests.get(query_url)
```

```python
r.status_code
```

```python
# Load the JSON data to a dictionary
j = json.loads(r.content)
```

```python
# This is quite a flat JSON structure, so can be loaded into a DataFrame
df = pd.DataFrame(j['StudyFieldsResponse']['StudyFields'])
```

```python
def de_list(input_field):
    if isinstance(input_field, list):
        if len(input_field) == 0:
            return None
        elif len(input_field) == 1:
            return input_field[0]
        else:
            return '; '.join(input_field)
    else:
        return input_field
```

```python
for c in df.columns:
    df[c] = df[c].apply(de_list)
```

```python
df['StudyFirstPostDate'] = pd.to_datetime(df.StudyFirstPostDate)
df = df.sort_values(by='StudyFirstPostDate', ascending=False)
```

```python
df[df.StudyType == 'Interventional'].head(100)
```

```python
timestamp = datetime.datetime.now().date().isoformat()

# Write all results
df.to_csv(f'immunotherapy_clinical_trials_{timestamp}.csv', index=False)

# Write just interventional trials
df[df.StudyType ==
   'Interventional'].to_csv(f'immunotheray_interventional_clinical_trials_{timestamp}.csv',
                                            index=False)
```

```python
#Separate name ()

targettherapytrials=pd.DataFrame()

targettherapy.columns = ['index','Names','Cancer']
mytherapynames=[]
for name in targettherapy.Names:
  name=name[:-1]
  x = name.split(' (')
  if(len(x)>1):
    mytherapynames.append(x.pop())
    mytherapynames.append(x.pop())
mytherapynames
mytherapynames=set(mytherapynames)
```

```python
targettherapy.columns = ['Name','Cancer']

targettherapytrials = pd.DataFrame(columns=['Ingredient','Name','Cancer'])

for row in targettherapy.Name:
    targettherapytrials["Ingredient"] = targettherapy["Name"].apply(lambda x: x.split('(')[0])
    targettherapytrials['Name'] = targettherapy["Name"].apply(lambda x: x.split('(')[1])
targettherapytrials['Cancer'] = targettherapy['Cancer']

targettherapytrials['Name'] = targettherapytrials['Name'].str[:-1]

targettherapytrials
```

```python
for n in mytherapynames:
```

```python
#Loop through both names in the API
from datetime import datetime
import numpy as np

a = targettherapytrials['Name'].to_list()
b = targettherapytrials['Ingredient'].to_list()
joinedlist = a + b

newdf = pd.DataFrame()
for n in joinedlist:
  immunotherapy_BASE_URL = 'https://clinicaltrials.gov/api/query/study_fields?expr="'+n+'"&max_rnk=1000&fmt=json'
  query_url = f'{immunotherapy_BASE_URL}&fields={",".join(cdc_extract_fields)}'
  r = requests.get(query_url)
  j = json.loads(r.content)

  if(j['StudyFieldsResponse']['NStudiesFound'] > 0):
    myTrial = pd.DataFrame(j['StudyFieldsResponse']['StudyFields'])
    for c in myTrial.columns:
      myTrial[c] = myTrial[c].apply(de_list)
      myTrial['StudyFirstPostDate'] = pd.to_datetime(myTrial.StudyFirstPostDate, format="['%B %d, %Y']")
      myTrial = myTrial.sort_values(by='StudyFirstPostDate', ascending=False)

    rowcount = myTrial.shape[0]
    newlist = n
    myTriallist = np.repeat(newlist,rowcount)
    myTrial["Name Query"] = myTriallist

    newdf = pd.concat([newdf, myTrial])
```

```python
print(newdf)
olddf= newdf
olddf['Rank'] = olddf['Rank'].astype(str)

g = olddf.groupby("OfficialTitle")
x = g.agg("first")
x.update(g.agg({"Name Query": ", ".join, "Rank": ", ".join}))
x["Name Query"] = x["Name Query"].apply(lambda x: ', '.join(list(set([y.strip().lower() for y in str(x).split(",")]))))
x = x.reset_index()
```
